```python
# Importing libraries
import cv2
import time
import numpy as np

fourcc= cv2.VideoWriter_fourcc(*'XVID')
out=cv2.VideoWriter('output.avi',fourcc,20.0, (640,480))


cap=cv2.VideoCapture(0) 
time.sleep(5)           #Sleeping for 5 seconds when the webcam starts
count=0
background=0

        #Capturing Video from Web Cam
for i in range(60):                       #60 iterations for this for loop
    ret,background=cap.read()             #reading the input
background=np.flip(background, axis=1)


while (cap.isOpened()):                   #this while loop runs till the webcam is kept open
    ret,img=cap.read()                    #Capturing again the video on which the operation is to be performed
    
    if not ret:
        break
        
    count+=1
    img=np.flip(img,axis=1)
    hsv=cv2.cvtColor(img, cv2.COLOR_BGR2HSV)  #converting BGR to HSV(Heu, Saturation, Value)
    
    #this code works for red cloak only
    #to make it work for different color of cloak change the hue value and color in the array
    lower=np.array([0,120,70])
    upper=np.array([9,255,255])
    mask1=cv2.inRange(hsv,lower,upper)
    
    #value of red color code in the array
    lower=np.array([171,120,70])           
    upper=np.array([180,255,255])
    mask2=cv2.inRange(hsv,lower,upper)
    
    mask1=mask1+mask2                        #Shades of colors stored in mask1 
    
    mask1=cv2.morphologyEx(mask1, cv2.MORPH_OPEN,np.ones((3,3),np.uint8), iterations=2) #Removing noise using MORPH_OPEN   
    mask1=cv2.morphologyEx(mask1, cv2.MORPH_DILATE,np.ones((3,3),np.uint8)) ##Smoothing image using MORPH_DILATE 

    mask2=cv2.bitwise_not(mask1)
    
    res1=cv2.bitwise_and(img,img,mask=mask2)                 #dividing cloak and background in two parts for color
    res2=cv2.bitwise_and(background,background,mask=mask1)   #replacing cloak with background
    
    finalOutput=cv2.addWeighted(res1,1,res2,1,0)
    
    out.write(finalOutput)
    cv2.imshow("InvisibleCloak",finalOutput)
    c = cv2.waitKey(10)
    if c == 27:                                              #press esc key to close the window
        break
    
cap.release()
out.release()
cv2.destroyAllWindows()
```
